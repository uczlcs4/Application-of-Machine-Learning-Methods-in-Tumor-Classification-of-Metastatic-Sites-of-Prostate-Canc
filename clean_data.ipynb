```python
####cleaning data 
##step1:  Remove any genes with gene expression values equals to 
##0 for all patients.
##step1 remove gene: count_step1
count_step1=0
gene_removed_step1=[]
index_lst_step1=[]
for i in range(len(mRNA_231_chen_data)):
    if all(mRNA_231_chen_data.iloc[i,1:]==0):
        count_step1=count_step1+1
        gene_removed_step1.append(mRNA_231_chen_data.iloc[i,0])
        index_lst_step1.append(i)

```
